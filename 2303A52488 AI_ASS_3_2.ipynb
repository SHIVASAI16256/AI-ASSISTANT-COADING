```python
def calculator(num1, num2, operation):
    """
    Perform basic arithmetic operations on two numbers.
    
    Args:
        num1: First number
        num2: Second number
        operation: String indicating the operation ('+', '-', '*', '/', '//')
    
    Returns:
        The result of the operation
    """
    if operation == '+':
        return num1 + num2
    elif operation == '-':
        return num1 - num2
    elif operation == '*':
        return num1 * num2
    elif operation == '/':
        if num2 == 0:
            return "Error: Division by zero"
        return num1 / num2
    elif operation == '//':
        if num2 == 0:
            return "Error: Division by zero"
        return num1 // num2
    else:
        return "Error: Invalid operation"

# Example usage
print(calculator(10, 5, '+'))   # 15
print(calculator(10, 5, '-'))   # 5
print(calculator(10, 5, '*'))   # 50
print(calculator(10, 5, '/'))   # 2.0
print(calculator(10, 3, '//'))  # 3
```

```python
def calculator(num1, num2, operator):
    """
    Simple calculator function performing basic arithmetic.

    Parameters:
    - num1: first numeric operand
    - num2: second numeric operand
    - operator: one of '+', '-', '*', '/' indicating the operation

    Returns:
    - The result of the arithmetic operation, or an error message for invalid input.

    Comments explaining each operation:
    - Addition ('+'): returns the sum of num1 and num2.
    - Subtraction ('-'): returns the difference num1 - num2.
    - Multiplication ('*'): returns the product of num1 and num2.
    - Division ('/'): returns num1 / num2; returns an error message if num2 is zero.
    """
    # Addition: return sum
    if operator == '+':
        return num1 + num2

    # Subtraction: return difference
    if operator == '-':
        return num1 - num2

    # Multiplication: return product
    if operator == '*':
        return num1 * num2

    # Division: handle division by zero
    if operator == '/':
        if num2 == 0:
            return "Error: Division by zero"
        return num1 / num2

    # Invalid operator
    return "Error: Invalid operator. Use '+', '-', '*', or '/'."

# Usage examples
# Example: calculator(10, 5, '+') should return 15
print(calculator(10, 5, '+'))   # 15
print(calculator(10, 5, '-'))   # 5
print(calculator(10, 5, '*'))   # 50
print(calculator(10, 5, '/'))   # 2.0
print(calculator(10, 0, '/'))   # Error: Division by zero
```

```python
def sort_marks(marks, descending=False):
    """
    Sort a list of student marks.

    Parameters:
    - marks: list of numeric marks (e.g., [78, 92, 65])
    - descending: if True, sort from highest to lowest; otherwise lowest to highest

    Returns:
    - A new list containing the sorted marks.

    Comments:
    - Uses Python's built-in `sorted()` for reliable, stable sorting.
    - Handles duplicate marks naturally (they remain in relative order).
    """
    # Return a sorted copy of the list; `reverse` controls descending order
    return sorted(marks, reverse=descending)

# Examples
marks = [78, 92, 65, 88, 92]
print("Original marks:", marks)
print("Sorted ascending:", sort_marks(marks))        # [65, 78, 88, 92, 92]
print("Sorted descending:", sort_marks(marks, descending=True))  # [92, 92, 88, 78, 65]
```

```python
import math

def is_prime(n):
    """
    Return True if n is prime, otherwise False.

    Efficient checks and edge handling:
    - Numbers <= 1 (including negatives and 1) are not prime.
    - 2 and 3 are prime.
    - Even numbers > 2 are not prime.
    - Test odd divisors up to the integer square root using `math.isqrt`.
    """
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False
    limit = math.isqrt(n)
    for i in range(3, limit + 1, 2):
        if n % i == 0:
            return False
    return True

# Examples (few-shot):
# 2 -> True, 3 -> True, 4 -> False, 1 -> False, 17 -> True, -5 -> False
tests = [2, 3, 4, 1, 17, -5]
for t in tests:
    print(f"is_prime({t}) ->", is_prime(t))
```
