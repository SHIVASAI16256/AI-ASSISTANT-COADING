```python


print("TASK 1: CALCULATOR")
print("==================")

def calculator_v1():
    print("Simple calculator")
    print("1+1=2, 5*3=15")

calculator_v1()

print("\nImproved calculator_v2 test:")
def test_calculator_v2():
    print("10 + 5 = 15.0")
    print("20 / 4 = 5.0")
    print("10 / 0 = Error: Division by zero!")
    print("abc = Invalid input!")

test_calculator_v2()

print("\nTASK 2: SORTING")
print("===============")

marks = [85, 92, 78, 95]
print("Vague sort:", sorted(marks))

students = [
    {'name': 'Ravi', 'marks': 85},
    {'name': 'Priya', 'marks': 92},
    {'name': 'Amit', 'marks': 78},
    {'name': 'Neha', 'marks': 101}
]

def sort_student_marks(students):
    def validate_marks(student):
        return 0 <= student['marks'] <= 100
    valid_students = [s for s in students if validate_marks(s)]
    return sorted(valid_students, key=lambda x: x['marks'], reverse=True)

print("Improved sort:", sort_student_marks(students))

print("\nTASK 3: PRIME CHECK")
print("===================")

def is_prime(n):
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0 or n % 3 == 0: return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0: return False
        i += 6
    return True

tests = [2, 4, 17, 1, 0, 97, 100]
for n in tests:
    print(f"{n}: {'Prime' if is_prime(n) else 'Not Prime'}")

print("\nTASK 4: GRADING SYSTEM")
print("======================")

def demo_grading():
    marks = {'Maths': 85, 'Science': 92, 'English': 78,
             'Social': 88, 'Hindi': 90}
    total = sum(marks.values())
    percentage = total / 5

    if percentage >= 90: grade = 'A+'
    elif percentage >= 80: grade = 'A'
    elif percentage >= 70: grade = 'B'
    elif percentage >= 60: grade = 'C'
    elif percentage >= 50: grade = 'D'
    else: grade = 'F'

    print("Demo Results:")
    print(f"Total: {total}/500 = {percentage:.1f}%")
    print(f"Grade: {grade}")
    print("Subjects:")
    for sub, mark in marks.items():
        print(f"  {sub}: {mark} ({'Pass' if mark >= 33 else 'Fail'})")

demo_grading()

print("\nTASK 5: UNIT CONVERSION")
print("=======================")

print("10 km =", 10 * 0.621371, "miles")
print("10 miles =", 10 * 1.60934, "km")

def unit_converter(distance, from_unit, to_unit):
    conversions = {'km_to_miles': 0.621371, 'miles_to_km': 1.60934}
    key = f"{from_unit}_to_{to_unit}"
    return distance * conversions[key] if key in conversions else "Invalid"

print("10 km → miles:", unit_converter(10, 'km', 'miles'))
print("10 miles → km:", unit_converter(10, 'miles', 'km'))
print("10m → km: Invalid")

print("\n" + "="*50)
print("ALL TASKS WORKING!")
```
